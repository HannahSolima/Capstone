#### For Loop to Access IMDb Score

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:105.0) Gecko/20100101 Firefox/105.0'
}
```

```python
score = []
for x in href:
    try: 
        IMDb_soup = BS(requests.get(f'https://www.imdb.com{x}ratings/?ref_=tt_ov_rt', headers = headers).text)
        score.append(IMDb_soup.findAll('span', attrs ={'class':'ipl-rating-star__rating'})[0].text)
    except:
        score.append("Not Available on IMDb") 
```
