#### Retrieving Original Language

```python
org_lang = []
for x in titles_clean:
    try: 
        IMDb_soup = BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
        parent = IMDb_soup.find('ul', attrs={'class': 'content-meta info'})
        children = parent.findChildren("li" , recursive=False)
        len_child = len(children)
        for y in range(0,len_child):
            item_child = children[y].get_text(strip=True).split(':')[0]
            if item_child.endswith('Language'):
                org_lang.append(children[y].get_text(strip=True).split(':')[-1])
            else:
                continue
    except: 
        org_lang.append("Not Available on IMDb")
```

#### US Box Office (gross amount)

```python
box = []
counter = 0
for x in titles_clean[0:10]:
    try: 
        IMDb_soup = BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
        print(counter)
        parent = IMDb_soup.find('ul', attrs={'class': 'content-meta info'})
        children = parent.findChildren("li" , recursive=False)
        len_child = len(children)
        counter += 1
        for y in range(0,len_child): 
            item_child = children[y].get_text(strip=True).split(':')[0]
            if item_child.endswith("USA)"): 
                box.append(children[y].get_text(strip=True).split(':')[-1])  
            else:
                continue
    except: 
        box.append("Not Available on IMDb")
```

#### Distributor

```python
dist = []
for x in titles_clean:
    try: 
        IMDb_soup = BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
        dist.append(IMDb_soup.findAll('div', attrs = {'class':'meta-value'})[10].text.replace('\n','').replace(' ',''))
    except:
        dist.append("Not Available on IMDb")
```
