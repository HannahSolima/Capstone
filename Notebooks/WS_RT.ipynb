#### Retrieving Original Language

```python
#big shoutout to ROB who helped me with this solution and explained to me exactly what was going on and why it works - you rock
org_lang = []
counter = 0
for x in titles_clean:
    IMDb_soup =  BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
    parent = IMDb_soup.find('ul', attrs={'class': 'content-meta info'})
    children = parent.findChildren("li" , recursive=False)
    i = len(children)
    print(counter)
    list1 = []
    index = 0
    for x in children:
        list1.append(children[index].get_text(strip=True).split(':')[0])
        index += 1
    if "Original Language" in list1:
        for x in range(0,i):
            y = children[x].get_text(strip=True).split(':')[0]
            if y.endswith('Language'):
                    org_lang.append(children[x].get_text(strip=True).split(':')[-1])
            else:
                continue
    else:
        org_lang.append("Not Available")
    counter += 1
```

#### US Box Office (gross amount)

```python
box = []
counter = 0
for x in titles_clean:
    IMDb_soup =  BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
    parent = IMDb_soup.find('ul', attrs={'class': 'content-meta info'})
    children = parent.findChildren("li" , recursive=False)
    i = len(children)
    print(counter)
    list1 = []
    index = 0
    for x in children:
        list1.append(children[index].get_text(strip=True).split(':')[0])
        index += 1
    if "Box Office (Gross USA)" in list1:
        for x in range(0,i):
            y = children[x].get_text(strip=True).split(':')[0]
            if y.endswith("USA)"):
                    box.append(children[x].get_text(strip=True).split(':')[-1])
            else:
                continue
    else:
        box.append("Not Available")
    counter += 1
```

#### Distributor

```python
dist = []
counter = 0
for x in titles_clean:
    IMDb_soup =  BS(requests.get(f'https://www.rottentomatoes.com/m/{x}', headers = headers).text)
    parent = IMDb_soup.find('ul', attrs={'class': 'content-meta info'})
    children = parent.findChildren("li" , recursive=False)
    i = len(children)
    print(counter)
    list1 = []
    index = 0
    for x in children:
        list1.append(children[index].get_text(strip=True).split(':')[0])
        index += 1
    if "Distributor" in list1:
        for x in range(0,i):
            y = children[x].get_text(strip=True).split(':')[0]
            if y.endswith('Distributor'):
                    dist.append(children[x].get_text(strip=True).split(':')[-1])
            else:
                continue
    else:
        dist.append("Not Available")
    counter += 1
```
